```python
# set up imports
import numpy as np
import torch
import statsmodels.nonparametric.smoothers_lowess
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
def power_utlity(x, alpha=0.1, beta=0.1, lamb=0.2):
    utility = x.copy()
    utility[x>=0.] = x[x>=0.]**(alpha)
    utility[x<0.] = -lamb*((-x[x<0.])**(-beta))
    return utility 
```

```python
def general_power_utlity(x, alpha=1.3, beta=0.5, gamma=1.5, lamb=0.1, delta=1.):
    utility = x.copy()
    utility[x>=0.] = beta*(x[x>=0.]**alpha)
    utility[x<0.] = -lamb*((-delta*x[x<0.])**(gamma))
    return utility 
```

```python
values = np.arange(-1000,1000)
utilities = power_utlity(values)
```

```python
plt.plot(values, utilities)
```

```python
values = np.arange(-1000,1000)
utilities = general_power_utlity(values)


fig, ax = plt.subplots(1, 1, figsize=(5, 5), dpi=100)
ax.plot(values, utilities, lw=2, c='k')
ax.set_xlabel(xlabel='values, x', fontsize=15)
ax.set_ylabel(ylabel='Utility, u(x)', fontsize=15)
ax.set_title(label="General Power Utility", fontsize=20)
plt.show()
```

```python
def kahneman_tversky(p, alpha=0.5):
    return p**alpha/(p**alpha + (1-p)**alpha)**(1/alpha)
```

```python
probs = np.arange(0,1., 0.01)
utilities = kahneman_tversky(probs)


fig, ax = plt.subplots(1, 1, figsize=(5, 5), dpi=100)
ax.plot(probs, utilities, lw=2, c='k')
ax.set_ylabel(ylabel='Decision weights, $\pi(p)$', fontsize=15)
ax.set_xlabel(xlabel='probabilities, p', fontsize=15)
ax.set_title(label="Kahneman-Tversky", fontsize=20)
plt.show()
```
